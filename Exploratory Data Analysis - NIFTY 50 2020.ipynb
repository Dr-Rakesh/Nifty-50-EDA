# Step 3: Perform exploratory analysis & visualization

```python
#creating a copy data frame to edit and perform exploratory analysis
axis_data=axisbank_data
bajaj_data=bajajfinsv_data
baj_data=bajfinance_data
hdfc1_data=hdfc_data
hdfcbk_data=hdfcbank_data
icici_data=icicibank_data
indus_data=indusindbk_data
kotak_data=kotakbank_data
sbi_data=sbibank_data
```

```python
#Extracting data year wise 
axis_data['year'] = pd.DatetimeIndex(axisbank_data.Date).year
bajaj_data['year'] = pd.DatetimeIndex(bajajfinsv_data.Date).year
baj_data['year'] = pd.DatetimeIndex(bajfinance_data.Date).year
hdfc1_data['year'] = pd.DatetimeIndex(hdfc_data.Date).year
hdfcbk_data['year'] = pd.DatetimeIndex(hdfcbank_data.Date).year
icici_data['year'] = pd.DatetimeIndex(icicibank_data.Date).year
indus_data['year'] = pd.DatetimeIndex(indusindbk_data.Date).year
kotak_data['year'] = pd.DatetimeIndex(kotakbank_data.Date).year
sbi_data['year'] = pd.DatetimeIndex(sbibank_data.Date).year
```

```python
#Sorting data year wise
axis_data_2020=axis_data[axis_data.year == 2020]
bajaj_data_2020=bajaj_data[bajaj_data.year == 2020]
baj_data_2020=baj_data[baj_data.year == 2020]
hdfc1_data_2020=hdfc1_data[hdfc1_data.year == 2020]
hdfcbk_data_2020=hdfcbk_data[hdfcbk_data.year == 2020]
icici_data_2020=icici_data[icici_data.year == 2020]
indus_data_2020=indus_data[indus_data.year == 2020]
kotak_data_2020=kotak_data[kotak_data.year == 2020]
sbi_data_2020=sbi_data[sbi_data.year == 2020]
```

```python
axis_data_2020.describe()
```

```python
df = [axis_data_2020, bajaj_data_2020, baj_data_2020, hdfc1_data_2020, hdfcbk_data_2020, icici_data_2020, indus_data_2020, kotak_data_2020, sbi_data_2020]
```

```python
#Creating a file comprising of all the data
result = pd.concat(df, keys=['Date', 'Symbol', 'Series', 'Prev Close', 'Open', 'High', 'Low', 'Last',
       'Close', 'VWAP', 'Volume', 'Turnover', 'Trades', 'Deliverable Volume',
       '%Deliverble'])
```

```python
result[:5]
```

```python
result.info()
```

```python
#finding daily open to close difference
result['Difference'] = result.Open - result.Close
```

```python
#Calculation of daily change
result['Percentage_diff']= (((result.Open - result.Close)*100)/result.Open)
```

```python
result
```

```python
result['Month']=pd.DatetimeIndex(result.Date).month
```

```python
#Finding maximum capturable change
result['Percent_daymove']= (((result.High - result.Low)*100)/result.High)
```

```python
#Finding over night after market hours trading activity
result['Percent_overnightmove']= (((result['Prev Close'] - result.Open)*100)/result['Prev Close'])
```

```python
result
```

```python
#After all the analysis segregating the data company wise
axis_df=result[result.Symbol == 'AXISBANK']
bajaj_df=result[result.Symbol == 'BAJAJFINSV']
baj_df=result[result.Symbol == 'BAJFINANCE']
hdfc1_df=result[result.Symbol == 'HDFC']
hdfc_df=result[result.Symbol == 'HDFCBANK']
icici_df=result[result.Symbol == 'ICICIBANK']
indus_df=result[result.Symbol == 'INDUSINDBK']
kotak_df=result[result.Symbol == 'KOTAKBANK']
sbi_df=result[result.Symbol == 'SBIN']
```

## Q1. What is the day trade profit margin for the year 2021?

```python
Percentageprofit_anually=[axis_df.Percentage_diff.sum(), bajaj_df.Percentage_diff.sum(), baj_df.Percentage_diff.sum(), hdfc1_df.Percentage_diff.sum(), hdfc_df.Percentage_diff.sum(), icici_df.Percentage_diff.sum(), indus_df.Percentage_diff.sum(), kotak_df.Percentage_diff.sum(), sbi_df.Percentage_diff.sum()]
```

```python
Percentageprofit_anually
```

```python
Bank = ['Axix bank', 'Bajaj finsv', 'Baj Finance', 'HDFC', 'HDFC Bank', 'ICICI Bank', 'INDUS Bank', 'Kotak Bank', 'SBI']
```

```python
plt.figure(figsize=(24, 10))
plt.bar(Bank, Percentageprofit_anually)
plt.xlabel("Bank")
plt.ylabel("Cummulative Percentage profit")
plt.title("Annual Profit/Loss")
plt.show()
```

## Q2. What is the possible maximum day trade profit margin for the year 2021?

```python
Percent_daymove_anually=[axis_df.Percent_daymove.sum(), bajaj_df.Percent_daymove.sum(), baj_df.Percent_daymove.sum(), hdfc1_df.Percent_daymove.sum(), hdfc_df.Percent_daymove.sum(), icici_df.Percent_daymove.sum(), indus_df.Percent_daymove.sum(), kotak_df.Percent_daymove.sum(), sbi_df.Percent_daymove.sum()]
```

```python
plt.figure(figsize=(24, 10))
plt.bar(Bank, Percent_daymove_anually)
plt.xlabel("Bank")
plt.ylabel("Possible Percentage profit")
plt.title("Annual Profit/Loss")
plt.show()
```

## Q3. Is over night market trading helpful?

```python
Percent_overnight_anually=[axis_df.Percent_overnightmove.sum(), bajaj_df.Percent_overnightmove.sum(), baj_df.Percent_overnightmove.sum(), hdfc1_df.Percent_overnightmove.sum(), hdfc_df.Percent_overnightmove.sum(), icici_df.Percent_overnightmove.sum(), indus_df.Percent_overnightmove.sum(), kotak_df.Percent_overnightmove.sum(), sbi_df.Percent_overnightmove.sum()]
```

## Q4. What is the risk for having both intraday and overnight trade for the year 2021?

```python
difference = [axis_df.Percentage_diff.sum()+axis_df.Percent_overnightmove.sum(), 
              bajaj_df.Percentage_diff.sum()+bajaj_df.Percent_overnightmove.sum(),
              baj_df.Percentage_diff.sum()+baj_df.Percent_overnightmove.sum(), 
              hdfc1_df.Percentage_diff.sum()+hdfc1_df.Percent_overnightmove.sum(), 
              hdfc_df.Percentage_diff.sum()+hdfc_df.Percent_overnightmove.sum(), 
              icici_df.Percentage_diff.sum()+icici_df.Percent_overnightmove.sum(), 
              indus_df.Percentage_diff.sum()+indus_df.Percent_overnightmove.sum(), 
              kotak_df.Percentage_diff.sum()+kotak_df.Percent_overnightmove.sum(), 
              sbi_df.Percentage_diff.sum()+sbi_df.Percent_overnightmove.sum()]
```

```python
plt.figure(figsize=(24, 10))
plt.bar(Bank,difference)
plt.xlabel("Bank")
plt.ylabel("Overnight Percentage profit")
plt.title("Overnight Hero/Zero")
plt.show()
```
